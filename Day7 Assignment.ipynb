2. Take a list of tuple as shown below.
[(1,2), (3,4), (5,6),(4,5)]
 
Make a new list which contains the sum of the number of tuples.
For example
Input
[(1,2), (3,4), (5,6)]
 
Output
 
[3, 7, 11]

```
list1=[(1,2),(3,4),(5,6),(4,5)]
list2=[]
for each in range(0,len(list1)):
    a,b=list1[each]
    list2.append(a+b)
print(" sum of tuple:",list2)
```
